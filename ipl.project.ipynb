```python
# to create a new column in extrect data year from date column
match_data['seaborn']=pd.DatetimeIndex(match_data['date']).year
```

```python
match_data.head(2)
```

```python
# number of matches count in per year and rename the column
match_per_seaon=match_data.groupby(['seaborn'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_seaon
```

```python
# to plot the graph 
sns.countplot(match_data['seaborn'])
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('season',fontsize=10)
plt.ylabel('count',fontsize=10)
plt.title('total matches played in each season',fontsize=10,fontweight='bold')
```

```python
# to join the to dataset
season_data=match_data[['id','seaborn']].merge(ball_data,left_on='id',right_on='id',how='left').drop('id',axis=1)
season_data.head(1)
```

```python
# plot the graph
season=season_data.groupby(['seaborn'])['total_runs'].sum().reset_index()
p=season.set_index('seaborn')
ax=plt.axes()
ax.set(facecolor='black')
sns.lineplot(data=p,palette='magma')
plt.title('total runs in each session',fontsize=12,fontweight='bold')
```

```python
match_data.head(1)
```

```python
toss=match_data['toss_winner'].value_counts()
ax=plt.axes()
ax.set(facecolor='grey')
sns.set(rc={'figure.figsize':(10,15)},style='darkgrid')
ax.set_title('no fo tosses win by each team',fontsize=15,fontweight='bold')
sns.barplot(y=toss.index,x=toss,orient='h',palette='icefire',saturation=5)
plt.xlabel('no. of tosses win')
plt.ylabel('teams')
```

```python
ax=plt.axes()
ax.set(facecolor='black')
sns.countplot(x='seaborn',hue='toss_decision',data=match_data,palette='magma')
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n seaborn',fontsize=10)
plt.ylabel('count',fontsize=15)
plt.title('toss decision across season',fontsize=20,fontweight='bold')
```
